```python
!pip install rapidfuzz

import pandas as pd
from rapidfuzz import process, fuzz


# Clean the CPU names in both DataFrames
cpu_data['cpu_Name_Cleaned'] = cpu_data['cpu_Name'].astype(str).str.strip().str.upper()
data['CPU_Cleaned'] = data['CPU'].astype(str).str.strip().str.upper()

# Calculate composite score (normalized weighted average)
max_single = cpu_data['single_core_score'].max()
max_multi = cpu_data['multi_core_score'].max()

cpu_data['CPU_Composite_Score'] = (
    (cpu_data['single_core_score'] / max_single) * 60 +
    (cpu_data['multi_core_score'] / max_multi) * 40
)

# Create a mapping dictionary from cleaned names to composite scores
score_mapping = dict(zip(
    cpu_data['cpu_Name_Cleaned'],
    cpu_data['CPU_Composite_Score']
))

# Fuzzy match function
def get_composite_score(cpu_name):
    if pd.isna(cpu_name):
        return None
    # Find best match in cpu_data
    result = process.extractOne(
        cpu_name,
        list(score_mapping.keys()),
        scorer=fuzz.token_set_ratio,
        score_cutoff=80
    )
    return score_mapping[result[0]] if result else None

# Apply matching and add scores to main data
data['CPU_Composite_Score'] = data['CPU_Cleaned'].apply(get_composite_score)


print("Successfully appended CPU_Composite_Score to your data.csv file!")
```

```python
# GPU Composite Score with 35 Similarity Threshold
import pandas as pd
from rapidfuzz import process, fuzz

## 1. Data Preparation ##
# Clean up GPU data
gpu_data.drop(columns=['price', 'gpuValue', 'TDP', 'category', 'testDate'], inplace=True)

# Normalize and create composite score
max_G2D = gpu_data['G2Dmark'].max()
max_G3D = gpu_data['G3Dmark'].max()

gpu_data['gpuCompositeScore'] = (
    (gpu_data['G3Dmark'] / max_G3D) * 90 +  # 90% weight for 3D
    (gpu_data['G2Dmark'] / max_G2D) * 10     # 10% weight for 2D
)

## 2. Name Cleaning ##
gpu_data['gpuNameClean'] = (
    gpu_data['gpuName']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(r'[^A-Z0-9\s]', '', regex=True)  # Remove special chars
)

data['GPU_Cleaned'] = (
    data['Graphical Controller']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(r'[^A-Z0-9\s]', '', regex=True)
)

## 3. Fuzzy Matching with 35 Threshold ##
gpu_score_map = dict(zip(gpu_data['gpuNameClean'], gpu_data['gpuCompositeScore']))

def get_gpu_score(gpu_name):
    if pd.isna(gpu_name) or not gpu_name:
        return None

    # First try exact match
    if gpu_name in gpu_score_map:
        return gpu_score_map[gpu_name], 100  # Return score and match confidence

    # Then try fuzzy match with low threshold
    result = process.extractOne(
        gpu_name,
        list(gpu_score_map.keys()),
        scorer=fuzz.token_set_ratio,
        score_cutoff=35  # Your requested low threshold
    )

    if result:
        matched_name, score, _ = result
        return gpu_score_map[matched_name], score
    return None, 0

## 4. Apply Matching ##
# Apply matching and unpack results
data[['GPU_Composite_Score', 'GPU_Match_Confidence']] = data['GPU_Cleaned'].apply(
    lambda x: pd.Series(get_gpu_score(x))
)


data.drop(columns=['GPU_Cleaned'], inplace=True, errors='ignore')

weak_matches = data[data['GPU_Match_Confidence'] < 60]
if not weak_matches.empty:
    weak_matches.to_csv('weak_gpu_matches_review.csv', index=False)

data.drop(columns=['GPU_Match_Confidence'], inplace=True, errors='ignore')


gpu_median = gpu_data['gpuCompositeScore'].median()
data['GPU_Composite_Score'] = data['GPU_Composite_Score'].fillna(gpu_median)
```

```python
def compute_performance_score(row):
    score = 0
    score += row['RAM Capacity'] * 0.35
    score += row['CPU Clock'] * 0.2
    score += row['GPU Clock'] * 0.15
    score += row['CPU_Composite_Score'] * 0.3
    score += row['Memory Capacity'] * 0.10
    score += row['Nominal Battery Capacity'] * 0.10
    score += row['RAM Type Score']
    score += row['GPU_Composite_Score'] * 0.2
    return round(score, 2)


data['Performance_Class'] = data.apply(compute_performance_score, axis=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Keep original column names here
categorical_features = ['CPU', 'RAM Type', 'Graphical Controller']
numeric_features = ['RAM Capacity', 'Nominal Battery Capacity', 'Max. Charging Power',
                    'Memory Capacity', 'CPU Clock', 'GPU Clock',
                    'CPU_Composite_Score', 'GPU_Composite_Score']

```

```python
for col in numeric_features:
    if data[col].isna().sum() > 0:
        data[col].fillna(data[col].mean(), inplace=True)
```

```python
for col in categorical_features:
    if data[col].isna().sum() > 0:
        data[col].fillna('Unknown', inplace=True)
```

```python
# Preprocessing step
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', StandardScaler(), numeric_features)
    ]
)

# Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])



# Fit
y = data['Performance_Type']  # or 'Performance_Class', depending on your naming
X = data[categorical_features + numeric_features]
pipeline.fit(X, y)
```

```python
# Load uploaded JSON file
with open('/content/device_stats.json') as f:
    raw_data = json.load(f)

# Flatten it into a single row dict
device = {
    'cpu_identifier': raw_data['cpu']['identifier'],
    'ram_capacity': raw_data['raminfo']['ram_capacity'],
    'ram_type': raw_data['raminfo']['ram_type'],
    'nominal_capacity_mah': raw_data['battery']['nominal_capacity_mah'],
    'max_charging_voltage_mv': raw_data['battery']['max_charging_voltage_mv'],
    'gpu_renderer': raw_data['gpu']['renderer'],
    'memory_capacity_gb': raw_data['memory_capacity_gb'],
    'cpu_clock_m_hz': raw_data['cpu_clock_m_hz'],
    'gpu_clock_m_hz': raw_data['gpu_clock_m_hz'],
}

df = pd.DataFrame([device])
```

```python
def match_best(name, choices, scores, cutoff=80):
    match = process.extractOne(name, choices)
    if match:
        matched_name, match_score = match
        if match_score >= cutoff:
            score_index = choices.index(matched_name)
            return scores.iloc[score_index], matched_name
    return 0, None  # fallback if no good match
```

```python
# Clean input fields
df['cpu_clean'] = df['cpu_identifier'].astype(str).str.upper().str.strip()
df['gpu_clean'] = df['gpu_renderer'].astype(str).str.upper().str.replace(r'[^A-Z0-9 ]', '', regex=True)

cpu_data['cpu_Name_Cleaned'] = cpu_data['cpu_Name'].astype(str).str.upper().str.strip()
gpu_data['gpuName_Cleaned'] = gpu_data['gpuName'].astype(str).str.upper().str.replace(r'[^A-Z0-9 ]', '', regex=True)

# Match function
def match_best(name, choices, scores, cutoff=80):
    match = process.extractOne(name, choices, score_cutoff=cutoff)
    if match:
        score_index = choices.index(match[0])
        return scores.iloc[score_index]
    return 0  # fallback

# Apply matches
df['cpu_composite_score'] = df['cpu_clean'].apply(
    lambda x: match_best(x, cpu_data['cpu_Name_Cleaned'].tolist(), cpu_data['CPU_Composite_Score'])
)

df['gpu_composite_score'] = df['gpu_clean'].apply(
    lambda x: match_best(x, gpu_data['gpuName_Cleaned'].tolist(), gpu_data['gpuCompositeScore'])
)
```
